```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from matplotlib import rcParams
from matplotlib import font_manager


# 设置苹果系统的中文字体
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# 生成示例数据
np.random.seed(0)
X_single = np.random.rand(100, 1) * 10  # 单变量数据
y_single = 2 * X_single + 1 + np.random.randn(100, 1)  # 单变量线性关系

# 多变量数据
X_multi = np.random.rand(100, 2) * 10  # 两个自变量
y_multi = 3 * X_multi[:, 0] + 2 * X_multi[:, 1] + 5 + np.random.randn(100)  # 多变量线性关系

# 创建单变量线性回归模型
model_single = LinearRegression()
model_single.fit(X_single, y_single)
y_single_pred = model_single.predict(X_single)

# 创建多变量线性回归模型
model_multi = LinearRegression()
model_multi.fit(X_multi, y_multi)
y_multi_pred = model_multi.predict(X_multi)

# 绘制单变量线性回归图表
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_single, y_single, color='blue', label='实际值')
plt.plot(X_single, y_single_pred, color='red', label='预测值')
plt.xlabel('自变量 X')
plt.ylabel('因变量 y')
plt.title('单变量线性回归')
plt.legend()

# 绘制多变量线性回归图表
plt.subplot(1, 2, 2)
plt.scatter(X_multi[:, 0], y_multi, color='blue', label='实际值 - X1')
plt.scatter(X_multi[:, 1], y_multi, color='green', label='实际值 - X2')
plt.scatter(X_multi[:, 0], y_multi_pred, color='red', alpha=0.5, label='预测值 - X1')
plt.scatter(X_multi[:, 1], y_multi_pred, color='orange', alpha=0.5, label='预测值 - X2')
plt.xlabel('自变量 X1 和 X2')
plt.ylabel('因变量 y')
plt.title('多变量线性回归')
plt.legend()

plt.tight_layout()
plt.savefig('linear-regression.png', format='png', transparent=True)
plt.show()
```
