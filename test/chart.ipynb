```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib import rcParams
from matplotlib import font_manager


# 
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 

# Generate some data
np.random.seed(0)
data = np.random.randn(1000)

fig, axs = plt.subplots(2, 3, figsize=(18, 12))

# Histogram
axs[0, 0].hist(data, bins=30, alpha=0.7, color='blue')
axs[0, 0].set_title(' (Histogram)')
axs[0, 0].set_xlabel('')
axs[0, 0].set_ylabel('')
axs[0, 0].grid(True)

# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 2, 4, 6]

# Bar chart
axs[0, 1].bar(categories, values, color='orange')
axs[0, 1].set_title(' (Bar Chart)')
axs[0, 1].set_xlabel('')
axs[0, 1].set_ylabel('')
axs[0, 1].grid(True)

# Pie chart
axs[0, 2].pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(np.arange(len(categories))))
axs[0, 2].set_title(' (Pie Chart)')

# Box plot
axs[1, 0].boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='cyan'))
axs[1, 0].set_title(' (Box plot)')
axs[1, 0].set_xlabel('')
axs[1, 0].grid(True)

# Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
axs[1, 1].scatter(x, y, color='red', alpha=0.6)
axs[1, 1].set_title(' (Scatter plot)')
axs[1, 1].set_xlabel('X ')
axs[1, 1].set_ylabel('Y ')
axs[1, 1].grid(True)

# Remove the last subplot (bottom-right) as we only have 5 plots
fig.delaxes(axs[1, 2])

plt.tight_layout()
plt.savefig('chart.png', format='png', transparent=True)
plt.show()

```
