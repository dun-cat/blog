```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

from matplotlib import rcParams
from matplotlib import font_manager


# 设置苹果系统的中文字体
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题

# 示例数据
data = {
    '组别': ['A']*10 + ['B']*10 + ['C']*10,
    '成绩': [85, 87, 88, 90, 86, 89, 92, 91, 87, 88, 78, 79, 80, 83, 81, 82, 85, 86, 84, 83, 92, 94, 93, 95, 96, 97, 99, 98, 97, 96]
}
df = pd.DataFrame(data)

# 单因素方差分析
f_val, p_val = stats.f_oneway(df[df['组别'] == 'A']['成绩'],
                              df[df['组别'] == 'B']['成绩'],
                              df[df['组别'] == 'C']['成绩'])


# 绘制图表
plt.figure(figsize=(10, 6))

# 箱线图
df.boxplot(column='成绩', by='组别', grid=False)
plt.title('各组成绩的箱线图')
plt.suptitle('')
plt.xlabel('组别')
plt.ylabel('成绩')
plt.savefig('anova.svg', format='svg', transparent=True)
plt.show()

# 输出F值和p值
print(f'F 值: {f_val}')
print(f'p 值: {p_val}')
```
