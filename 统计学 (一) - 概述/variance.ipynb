```python

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

import pandas as pd


from matplotlib import rcParams
from matplotlib import font_manager


# 设置苹果系统的中文字体
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题


# 读取本地CSV文件
file_path = 'SOCR-HeightWeight.csv'  # 请将此路径替换为你的本地文件路径
data = pd.read_csv(file_path)

# 提取身高和体重数据
heights = data['Height(Inches)'] * 2.54
weights = data['Weight(Pounds)'] * 0.4535924

# 计算均值和标准差
mean_height = np.mean(heights)
std_dev_height = np.std(heights, ddof=1)
mean_weight = np.mean(weights)
std_dev_weight = np.std(weights, ddof=1)

# plt.figure(figsize=(12, 8))

# 绘制身高和体重的散点图
plt.scatter(heights, weights, color='blue', alpha=0.5, s=5,label='Height vs Weight')

# 标注均值线
plt.axvline(mean_height, color='red', linestyle='--', label=f'Mean Height: {mean_height:.2f}')
plt.axhline(mean_weight, color='orange', linestyle='--', label=f'Mean Weight: {mean_weight:.2f}')

# 绘制身高和体重的标准差区域
plt.fill_betweenx(np.arange(min(weights), max(weights)), mean_height - std_dev_height, mean_height + std_dev_height, color='red', alpha=0.1, label='±1 SD Height')
plt.fill_between(np.arange(min(heights), max(heights)), mean_weight - std_dev_weight, mean_weight + std_dev_weight, color='orange', alpha=0.1, label='±1 SD Weight')

# 添加图例
plt.legend()

# 添加标题和标签
plt.title('Scatter Plot of Height and Weight with Mean and Standard Deviation')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

plt.savefig('variance.png', format='png', transparent=True)

# 显示图表
plt.show()

# 输出均值和标准差
print(f'Mean Height: {mean_height:.2f}')
print(f'Standard Deviation of Height: {std_dev_height:.2f}')
print(f'Mean Weight: {mean_weight:.2f}')
print(f'Standard Deviation of Weight: {std_dev_weight:.2f}')

```
