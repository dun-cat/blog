```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
data = [1, 3, 4, 5, 7, 8, 9]


# 获取四分位数和中位数
q1 = np.percentile(data, 25)
median = np.percentile(data, 50)
q3 = np.percentile(data, 75)

# 计算 IQR
iqr = q3 - q1

# 绘制箱线图
fig, ax = plt.subplots()
box = ax.boxplot(data,vert=False, patch_artist=True)

# 修改箱子的背景色
for patch in box['boxes']:
    patch.set_facecolor('#f93a37')

# 标注 Q1, Q2, Q3 和 IQR
ax.text(q1, 1.05, f'Q1: {q1}', verticalalignment='center', fontsize=12, color='blue')
ax.text(median, 1.05, f'Q2: {median}', verticalalignment='center', fontsize=12, color='green')
ax.text(q3, 1.05, f'Q3: {q3}', verticalalignment='center', fontsize=12, color='red')
ax.text((q1 + q3) / 2, 0.88, f'IQR: {iqr}', horizontalalignment='center', fontsize=12, color='purple')

# 添加图表标题和标签
plt.title('IQR')
plt.xlabel('Value')

plt.savefig('iqr.svg', format='svg', transparent=True)

# 显示图表
plt.show()
```
