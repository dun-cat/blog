```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from matplotlib import rcParams
from matplotlib import font_manager


# 
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# 
np.random.seed(0)
X_single = np.random.rand(100, 1) * 10  # 
y_single = 2 * X_single + 1 + np.random.randn(100, 1)  # 

# 
X_multi = np.random.rand(100, 2) * 10  # 
y_multi = 3 * X_multi[:, 0] + 2 * X_multi[:, 1] + 5 + np.random.randn(100)  # 

# 
model_single = LinearRegression()
model_single.fit(X_single, y_single)
y_single_pred = model_single.predict(X_single)

# 
model_multi = LinearRegression()
model_multi.fit(X_multi, y_multi)
y_multi_pred = model_multi.predict(X_multi)

# 
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_single, y_single, color='blue', label='')
plt.plot(X_single, y_single_pred, color='red', label='')
plt.xlabel(' X')
plt.ylabel(' y')
plt.title('')
plt.legend()

# 
plt.subplot(1, 2, 2)
plt.scatter(X_multi[:, 0], y_multi, color='blue', label=' - X1')
plt.scatter(X_multi[:, 1], y_multi, color='green', label=' - X2')
plt.scatter(X_multi[:, 0], y_multi_pred, color='red', alpha=0.5, label=' - X1')
plt.scatter(X_multi[:, 1], y_multi_pred, color='orange', alpha=0.5, label=' - X2')
plt.xlabel(' X1  X2')
plt.ylabel(' y')
plt.title('')
plt.legend()

plt.tight_layout()
plt.savefig('linear-regression.png', format='png', transparent=True)
plt.show()
```
