```python
import matplotlib.pyplot as plt
import numpy as np

import scipy.stats as stats
from matplotlib import rcParams
from matplotlib import font_manager


# 
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 

# 
fig, ax = plt.subplots(4, 3, figsize=(15, 20))
ax = ax.flatten()

# x
x = np.linspace(-10, 10, 400)
x_positive = np.linspace(0.1, 10, 400)  # 
x_01 = np.linspace(-1, 1, 400)

# 
functions = [
    {"name": "", "func": lambda x: x, "domain": x},
    {"name": "", "func": lambda x: np.full_like(x, 2), "domain": x},
    {"name": "", "func": lambda x: 2 * x + 1, "domain": x},
    {"name": "", "func": lambda x: x**3 - 2*x**2 + x, "domain": x},
    {"name": "", "func": lambda x: 2**x, "domain": x},
    {"name": "", "func": lambda x: np.log(x), "domain": x_positive},
    {"name": "", "func": lambda x: np.sin(x), "domain": x},
    {"name": "", "func": lambda x: np.cos(x), "domain": x},
    {"name": "", "func": lambda x: np.tan(x), "domain": x},
    {"name": "", "func": lambda x: np.arcsin(np.clip(x, -1, 1)), "domain": x_01},
    {"name": "", "func": lambda x: np.arccos(np.clip(x, -1, 1)), "domain": x_01},
    {"name": "", "func": lambda x: np.arctan(x), "domain": x},
]

# 
for i, func in enumerate(functions):
    ax[i].plot(func["domain"], func["func"](func["domain"]))
    ax[i].set_title(func["name"])
    ax[i].axhline(0, color='black',linewidth=0.5)
    ax[i].axvline(0, color='black',linewidth=0.5)
    ax[i].grid(color = 'gray', linestyle = '--', linewidth = 0.5)

plt.tight_layout()

#  SVG 
plt.savefig('function_types.svg', format='svg', transparent=True)
plt.show()
```
