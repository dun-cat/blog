```python
import matplotlib.pyplot as plt
import numpy as np

import scipy.stats as stats
from matplotlib import rcParams
from matplotlib import font_manager


# 设置苹果系统的中文字体
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题

# 创建一个子图
fig, ax = plt.subplots(4, 3, figsize=(15, 20))
ax = ax.flatten()

# 定义x范围
x = np.linspace(-10, 10, 400)
x_positive = np.linspace(0.1, 10, 400)  # 用于对数函数
x_01 = np.linspace(-1, 1, 400)

# 定义不同的函数
functions = [
    {"name": "恒等函数", "func": lambda x: x, "domain": x},
    {"name": "常数函数", "func": lambda x: np.full_like(x, 2), "domain": x},
    {"name": "线性函数", "func": lambda x: 2 * x + 1, "domain": x},
    {"name": "多项式函数", "func": lambda x: x**3 - 2*x**2 + x, "domain": x},
    {"name": "指数函数", "func": lambda x: 2**x, "domain": x},
    {"name": "对数函数", "func": lambda x: np.log(x), "domain": x_positive},
    {"name": "正弦函数", "func": lambda x: np.sin(x), "domain": x},
    {"name": "余弦函数", "func": lambda x: np.cos(x), "domain": x},
    {"name": "正切函数", "func": lambda x: np.tan(x), "domain": x},
    {"name": "反正弦函数", "func": lambda x: np.arcsin(np.clip(x, -1, 1)), "domain": x_01},
    {"name": "反余弦函数", "func": lambda x: np.arccos(np.clip(x, -1, 1)), "domain": x_01},
    {"name": "反正切函数", "func": lambda x: np.arctan(x), "domain": x},
]

# 绘制所有函数
for i, func in enumerate(functions):
    ax[i].plot(func["domain"], func["func"](func["domain"]))
    ax[i].set_title(func["name"])
    ax[i].axhline(0, color='black',linewidth=0.5)
    ax[i].axvline(0, color='black',linewidth=0.5)
    ax[i].grid(color = 'gray', linestyle = '--', linewidth = 0.5)

plt.tight_layout()

# 保存为无背景的 SVG 格式
plt.savefig('function_types.svg', format='svg', transparent=True)
plt.show()
```
