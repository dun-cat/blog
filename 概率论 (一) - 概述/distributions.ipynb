```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib import rcParams
from matplotlib import font_manager


# 设置苹果系统的中文字体
rcParams['font.family'] = ['Hannotate SC','Hiragino Sans GB']
rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题

# 设置随机种子以确保结果可重复
np.random.seed(42)
```

二项分布

```python
# 二项分布
n, p = 10, 0.5
binom_rv = stats.binom(n, p)
x = np.arange(0, n + 1)
plt.bar(x, binom_rv.pmf(x), width=0.5, alpha=0.7, color='blue')
plt.title('二项分布 (Binomial Distribution)')
plt.xlabel('x')
plt.ylabel('概率')
plt.savefig('binomial_distribution.svg', format='svg', transparent=True)

plt.show()
```

几何分布

```python
# 几何分布
p = 0.5
geom_rv = stats.geom(p)
x = np.arange(1, 11)
plt.bar(x, geom_rv.pmf(x), width=0.5, alpha=0.7, color='green')
plt.title('几何分布 (Geometric Distribution)')
plt.xlabel('x')
plt.ylabel('概率')
plt.savefig('geometric_distribution.svg', format='svg', transparent=True)

plt.show()
```

泊松分布


```python
# 泊松分布
mu = 3
poisson_rv = stats.poisson(mu)
x = np.arange(0, 10)
plt.bar(x, poisson_rv.pmf(x), width=0.5, alpha=0.7, color='red')
plt.title('泊松分布 (Poisson Distribution)')
plt.xlabel('x')
plt.ylabel('概率')
plt.savefig('poisson_distribution.svg', format='svg', transparent=True)

plt.show()
```

正态分布


```python
# 正态分布
mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x, mu, sigma), color='purple')
plt.title('正态分布 (Normal Distribution)')
plt.xlabel('x')
plt.ylabel('概率密度')
plt.savefig('normal_distribution.svg', format='svg', transparent=True)

plt.show()
```

指数分布


```python
# 指数分布
scale = 1
x = np.linspace(0, 5, 1000)
plt.plot(x, stats.expon.pdf(x, scale=scale), color='orange')
plt.title('指数分布 (Exponential Distribution)')
plt.xlabel('x')
plt.ylabel('概率密度')
plt.savefig('exponential_distribution.svg', format='svg', transparent=True)

plt.show()
```

均匀分布


```python
# 均匀分布
loc, scale = 0, 1
x = np.linspace(-0.5, 1.5, 1000)
plt.plot(x, stats.uniform.pdf(x, loc, scale), color='brown')
plt.title('均匀分布 (Uniform Distribution)')
plt.xlabel('x')
plt.ylabel('概率密度')
plt.savefig('uniform_distribution.svg', format='svg', transparent=True)

plt.show()
```

伽马分布


```python
# 伽马分布
shape, scale = 2, 1
x = np.linspace(0, 10, 1000)
plt.plot(x, stats.gamma.pdf(x, shape, scale=scale), color='cyan')
plt.title('伽马分布 (Gamma Distribution)')
plt.xlabel('x')
plt.ylabel('概率密度')
plt.savefig('gamma_distribution.svg', format='svg', transparent=True)

plt.show()
```
