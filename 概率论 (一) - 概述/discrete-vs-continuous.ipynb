```python
# 设置随机种子以确保结果可重复
np.random.seed(42)

# 创建一个包含两个子图的图形对象，左右布局
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 绘制离散分布：泊松分布
mu = 3
poisson_rv = stats.poisson(mu)
x = np.arange(0, 10)
ax1.bar(x, poisson_rv.pmf(x), width=0.5, alpha=0.7, color='red')
ax1.set_title('离散随机变量（泊松分布）')
ax1.set_xlabel('x （随机变量）')
ax1.set_ylabel('概率')

# 绘制连续分布：正态分布
mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
ax2.plot(x, stats.norm.pdf(x, mu, sigma), color='purple')
ax2.set_title('连续随机变量（正态分布）')
ax2.set_xlabel('x（随机变量）')
ax2.set_ylabel('概率密度')

# 保存为无背景的 SVG 格式
plt.savefig('discrete_vs_continuous_distribution.svg', format='svg', transparent=True)
plt.show()
```
